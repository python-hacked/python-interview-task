```python
import cv2
import numpy
```

```python
face_cascade = cv2.CascadeClassifier('haarcascade_eye.xml') 
```

```python
# img = cv2.imread('boys.jpg')
cap = cv2.VideoCapture(0)
```

```python
while True:
    ret,img = cap.read()
    faces = face_cascade.detectMultiScale(img)


    print(faces)

    for (x, y, w, h)in faces:
        cv2.rectangle(img, (x,y), (x+w,y+h), (255,0,0),3)

    cv2.imshow('img1', img)
    if cv2.waitKey(1)==27:
        break
```

```python
cv2.waitKey(0)
cap.release()
```
